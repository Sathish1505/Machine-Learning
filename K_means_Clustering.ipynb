```python
#Call center data for clustering
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Call Center Data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['Experience in months'],df['Call Attended in a day'])
plt.show()
```

```python
from sklearn.cluster import KMeans
kmodel=KMeans(n_clusters=4)
```

```python
kmodel.fit(df)
```

```python
df['label']=kmodel.predict(df)
```

```python
kmodel.cluster_centers_
```

```python
df.head()
```

```python
plt.scatter(df['Experience in months'],df['Call Attended in a day'],c=df['label'])
plt.show()
```

```python
#Mall Data for clustering
```

```python
df=pd.read_csv('mall.csv')
```

```python
df.head()
```

```python
df1=df[['Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
df1.head()
```

```python
loss=[]
for i in range(1,12):
  kmodel=KMeans(n_clusters=i)

  kmodel.fit(df1)
  loss.append(kmodel.inertia_)
```

```python
plt.plot(range(1,12),loss)
plt.show()
```

```python
kmodel=KMeans(n_clusters=5)
kmodel.fit(df1)
```

```python
kmodel.predict(df1)
```

```python
df1['label']=kmodel.predict(df1)
```

```python
df1.head()
```

```python
kmodel.cluster_centers_
```

```python
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],c=df1['label'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
